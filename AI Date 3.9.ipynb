```python
import pandas as pd

import numpy as np

import seaborn as sns

import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import StandardScaler, LabelEncoder

from sklearn.neighbors import KNeighborsClassifier

from sklearn.datasets import load_breast_cancer

breast_cancer_data = load_breast_cancer()
```

```python
df_features = breast_cancer_data.feature_names

df_data = pd.DataFrame(breast_cancer_data.data, columns = df_features)

df_labels = pd.DataFrame(breast_cancer_data.target)
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_data, df_labels, stratify=df_labels, test_size=0.2, random_state=42)
```

```python
print("x_train.shape :", x_train.shape)

print("y_train.shape :", y_train.shape)

print("x_test.shape:", x_test.shape)

print("y_test.shape:", y_train.shape)
```

```python
# scaler

scaler = StandardScaler()

scaler.fit(x_train)

x_train_scaled = scaler.transform(x_train)

x_test_scaled = scaler.transform(x_test)

plt.scatter(x_train_scaled[:,0],x_train_scaled[:,1])
```

```python
mymodel = KNeighborsClassifier()

mymodel.fit(x_train_scaled, y_train)

prediction = mymodel.predict(x_test_scaled)

mymodel.score(x_test_scaled, y_test)
```
